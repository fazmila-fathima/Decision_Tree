Data Preprocessing

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
df = pd.read_csv('../input/500-person-gender-height-weight-bodymassindex/500_Person_Gender_Height_Weight_Index.csv')
```

```python
x= df.iloc[ : , 1: ]
y= df.iloc[ : , : 1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size= 0.20, random_state=0)
print(y_test)
print(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train = st_x.fit_transform(x_train)
x_test = st_x.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
Classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)
Classifier.fit(x_train, y_train)
```

Information Gain

```python
y_pred=Classifier.predict(x_test)
plt.plot(y_pred)
plt.show
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
plt.plot(cm)
plt.show
```

```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(Classifier,feature_names= x.columns, filled=True, class_names=['1','2','3','4','5','6','7'], out_file=None))
```
